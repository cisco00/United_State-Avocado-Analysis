```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

### Loading dataset into python environment

```python
df = pd.read_csv("/home/oem/Downloads/avocado.csv")
df.tail(10)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Average prices of organic and conventional Avocado across the region in the United State

```python
fig = px.bar(df, x = "region", y = "AveragePrice", color="type")
fig.show()
```

#### Converting date column into date and splitting it into day, month and year.

```python
df['date'] = pd.to_datetime(df['Date'])

df[["year", "month", "day"]] = df['Date'].str.split("-", expand=True)
```

#### Dropping redundant column

```python
df.drop(columns=['Date'], axis=0)
```

```python
df[['month', 'year', 'day']].head()
```

### Total volume of Avocado on a monthly basis

```python
group1 = df.groupby("month")['Total Volume'].count().plot()
group1
```

### Total volume of organic and conventional Avocado per month in the United State

```python
sns.barplot(data=df, x='month', y='Total Volume', hue='type');
```

### Total bags of avocado according based regions

```python
fig = px.bar(x=df['day'], y=df['Total Bags'], color=df['year'])
fig.show()
```

#### The monthly average price of organic avocado in different region 

```python
organic_df = df.loc[(df['type'] == 'organic')]
fig = px.bar(data_frame=organic_df, x='region', y='AveragePrice', color="month")
fig.show()
```

```python
organic_df = df.loc[(df['type'] == 'conventional')]
fig = px.bar(data_frame=organic_df, x='region', y='AveragePrice', color="month")
fig.show()
```

#### From the graph above we can categorically say that the prices of Avocado is based on the type of  avocado for instance the average price of organic avocado across the region differs by $100.
